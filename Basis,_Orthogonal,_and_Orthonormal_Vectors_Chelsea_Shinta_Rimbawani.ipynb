```python
import numpy as np
```

```python
i = np.array([1, 0])
i

j = np.array([0,1])
j # buat vector i dengan elemen (1,0) dan vector j dengan elemen (0,1)
```

```python
np.dot(i, j) # buktikan bahwa vector i dan vector j itu orthogonal
```
