Soal 2

```python
v = np.array([3, 4]) # buat vector v dengan elemen (3,4)

norm = np.linalg.norm(v) # hitung nilai L2 norm dari vector v

unit_v = v / norm

print("Vektor:", v)
print("Nilai norm:", norm)
print("Unit vector:", unit_v) # tampilkan unit vector dari vector v
print("Panjang unit vector:", np.linalg.norm(unit_v)) # hitung panjang unit vector dari vector v
```
